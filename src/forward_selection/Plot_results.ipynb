```python
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
mses = []

for i in range(48):    
    with open(f"src/metrics{i+1}.txt", "r") as f:
        l1 = f.readline()
        l2 = f.readline()
        mses.append(round(float(l2.strip("Best MSE: ").strip(", \n")),2))

min_index = 0
minlist = 100000000
for i in range(48):
    if mses[i] < minlist:
        minlist = mses[i]
        min_index = i
    
print("Best No. of Features: ", min_index+1)   
print("MSE: ", minlist)

fig = plt.figure()
plt.plot(mses)
plt.scatter(min_index, minlist, color="red")
plt.xlabel("No. of Features")
plt.ylabel("Validation Error (MSE)")
plt.show()

print("""For Features: 
hour
dayofweek
dayofyear
Madrid_t2m
holidays
Malaga_t2m
Malaga_d2m
weekofyear
hour3
Seville_t2m
Seville_tp
quarter
Zaragoza_t2m
Valencia_tp
Barcelona_tcc
dayofmonth
Madrid_tcc
Barcelona_i10fg
weekend
month
Zaragoza_d2m
Madrid_i10fg
Seville_tcc
Barcelona_tp
""")
```
